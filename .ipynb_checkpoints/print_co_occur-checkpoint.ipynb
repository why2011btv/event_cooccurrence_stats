```python
import csv
count = 0
with open('verb_cooccur_10.csv', 'w') as f:        
    with open('output0323.csv', 'r') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            if int(row[-1]) >= 10:
                count += 1
                f.write("%s,%s,%s\n"%(row[0], row[1], row[2]))
                
print(count)
# 2237243 NYT articles in total
```

```python
import os
def print_co_occur_verbs(key_word):
    command = "grep -w '^" + key_word + "' verb_cooccur_10.csv"
    result = os.popen(command).read()
    keyword_dict = {}
    for row in result.split("\n"):
        if row != '':
            row = row.split(",")
            keyword_dict[(row[0], row[1])] = int(row[2])

    sorted_dict = {k: v for k, v in sorted(keyword_dict.items(), key=lambda item: item[1], reverse=True)}
    for (key, value) in sorted_dict.items():
        print(key, value)
        
def print_co_occur(keyword_1, keyword_2):
    command = "grep -w '^" + keyword_1 + "," + keyword_2 + "' verb_cooccur_10.csv"
    result = os.popen(command).read()
    print(result.split(",")[2])
    
```

```python
print_co_occur("abduct", "drive")
```

```python
print_co_occur("clash", "wound")
```

```python
print_co_occur("attack", "injure")
```
